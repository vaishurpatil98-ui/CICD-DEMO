```python
import pandas as pd

from sklearn.feature_extraction.text import CountVectorizer

from sklearn.model_selection import train_test_split

from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score
```

```python
dataset = pd.read_csv("/content/spam1.csv", encoding="latin1")
```

```python
dataset.head()
```

```python
from numpy import vectorize
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(dataset["v2"])
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, dataset["v1"], test_size=0.2)
```

```python
model = MultinomialNB()
model.fit(x_train, y_train)
```

```python
yPred = model.predict(x_test)

# Evaluate the model

accuracy = accuracy_score(y_test, yPred)
print("Accuracy:", accuracy)
```

```python
# function to predict if a message is spam or not

def predictMessage(message):
  messageVector = vectorizer.transform([message])
  prediction_label = model.predict(messageVector)
  return 'Spam' if prediction_label[0] == 'spam' else 'Ham'

# get user input to perfect

userMessage = input("Enter a message: ")
message_result = predictMessage(userMessage)
print(message_result)
print(f'The message is: {message_result}')
```
